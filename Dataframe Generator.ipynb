## Analysis on Individual Lab Results

```python
LAB_THRESHOLD = 250000
LABS = []

for col in df.columns:
    if col.endswith("_ID") and "SUBJECT" not in col and "HADM" not in col: # all lab columns
        if len(df[df[col] == int(col[:-3])]) >= LAB_THRESHOLD:
            print("Lab:", col, "has", len(df[df[col] == int(col[:-3])]), "entries")
            LABS.append(col[:-3]+"_VALUENUM")
            LABS.append(col[:-3]+"_ABNORMAL_FLAG")

print("Labs we will use:", LABS)
print(len(LABS)//2)
```

```python
# Feature Set
VITALS = ["SUBJECT_ID", "HADM_ID", "TIME_BLOCKS_SINCE_ADMITTED", "DISCHARGED_WITHIN_BLOCKS","READMITTED_TIMES_REMAINING"]
FINAL_FRAME = VITALS + LABS
```

```python
patient_df = pd.read_csv("mimic-iii-clinical-database-1.4/ADMISSIONS.csv")
```

```python
patient_basics = patient_df[["HADM_ID", "LANGUAGE", "INSURANCE", "MARITAL_STATUS", "ETHNICITY"]]
```

```python
patient_basics["ETHNICITY"].unique() 
```

```python
for idx, elem in enumerate(patient_basics["INSURANCE"].unique()):
    patient_basics.loc[patient_basics["INSURANCE"] == elem, "INSURANCE"] = idx
```

```python
for idx, elem in enumerate(patient_basics["MARITAL_STATUS"].unique()):
    patient_basics.loc[patient_basics["MARITAL_STATUS"] == elem, "MARITAL_STATUS"] = idx
patient_basics["MARITAL_STATUS"] = patient_basics["MARITAL_STATUS"].fillna(6)
```

```python
for idx, elem in enumerate(patient_basics["LANGUAGE"].unique()):
    if elem == "ENGL":
        patient_basics.loc[patient_basics["LANGUAGE"] == elem, "LANGUAGE"] = 1
    elif elem == "SPAN":
        patient_basics.loc[patient_basics["LANGUAGE"] == elem, "LANGUAGE"] = 2
    elif elem == "CANT" or elem == "MAND":
        patient_basics.loc[patient_basics["LANGUAGE"] == elem, "LANGUAGE"] = 3
    else:
        patient_basics.loc[patient_basics["LANGUAGE"] == elem, "LANGUAGE"] = 0
        
patient_basics["LANGUAGE"] = patient_basics["LANGUAGE"].fillna(0)
```

```python
for idx, elem in enumerate(patient_basics["ETHNICITY"].unique()):
    if "WHITE" in elem or "PORTUGUESE" in elem:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 0
    elif "ASIAN" in elem:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 1
    elif "AFRICAN" in elem or "BLACK" in elem:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 2
    elif "HISPANIC" in elem or "SOUTH AMERICAN" in elem:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 3
    elif "MIDDLE EASTERN" in elem:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 4
    elif "NATIVE"in elem:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 5
    else:
        patient_basics.loc[patient_basics["ETHNICITY"] == elem, "ETHNICITY"] = 6
```
